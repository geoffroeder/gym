```python
import networkx as nx
import numpy as np
import gym
```

```python
class RandomAgent(object):
    """The world's simplest agent!"""
    def __init__(self, action_space):
        self.action_space = action_space

    def get_action(self, observation, reward, done):
        return self.action_space.sample()
```

```python
topology = [[1,2],[2,3],[3,4],[2,4]]
env = gym.make("GymternetIntranet-v0")
env.env.set_topology(topology)

obs = env.reset()
action_space = env.env.action_space
agent = RandomAgent(action_space)
reward = 0
done = False
iter = 0

while True:
   action = agent.get_action(obs, reward, done)
   obs, reward, done, info = env.step(action)
   iter += 1
   print("Epoch: {}; Reward: {}".format(iter, reward))
   if done:
      print("Total reward: {}".format(env.env.cumul_reward))
      env.reset()
      break
env.close()
```
